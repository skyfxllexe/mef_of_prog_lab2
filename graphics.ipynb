```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
def plot_times(file_name, legend, color):
    file_path = file_name

    x = []
    y = []

    with open(file_path, 'r') as file:
        for line in file:
            values = line.strip().split(',')
            x.append(int(values[0]))
            y.append(float(values[1]))
    plt.plot(x, y, marker='o', linestyle='-', color=color, label = legend)
    plt.legend()
    plt.ticklabel_format(axis='x', style='plain')

    # Показать график
```

```python
plot_times('times_rbtree.csv', "rbtree", "b")
plot_times('times_linear.csv', "linear", "g")
plot_times('times_binary.csv', "binary", "r")
plot_times('times_hash.csv', "hash", "y")
plot_times('times_multimap.csv', "multimap", "black")
plt.yscale('log')

# plt.show()
```
